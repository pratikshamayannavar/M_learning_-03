Data Preprocessing

```python
# separate the data and Label
X = wine_data.drop('quality',axis=1)
X.head()
```

```python
Y = wine_data['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
Y
```


Train & Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model Training:

Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
```

```python
# accuracy on test data
X_test_prediction = log_reg.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy : ', test_data_accuracy*100)
```

Decision Tree Classifier

```python
dtreec=DecisionTreeClassifier()
dtreec.fit(X_train,Y_train)
```

```python
# accuracy on test data
X_test_prediction = dtreec.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy: ', test_data_accuracy*100)
```

Random Forest Classifier

```python
Random_for_cls = RandomForestClassifier()
Random_for_cls.fit(X_train, Y_train)
```

```python
# accuracy on test data
X_test_prediction = Random_for_cls.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy : ', test_data_accuracy*100)
```

Random forest classifier gives the best accuracy of 94.0625

Building a Predictive System

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = Random_for_cls.predict(input_data_reshaped)
print(prediction)
if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
